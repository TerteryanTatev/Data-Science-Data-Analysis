```python
import pandas as pd
```

```python
df = pd.read_csv("/kaggle/input/netflix-data-cleaning-analysis-and-visualization/netflix1.csv")
df
```

```python
df.drop_duplicates(inplace=True)
print(df)
```

```python
df.isnull().sum()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])

df
```

```python
df['month_added'] =df['date_added'].dt.month_name()
df['year_added'] = pd.DatetimeIndex(df['date_added']).year
df['day_added'] = df['date_added'].dt.day_name()
df
```

```python
df.groupby("type").count()['show_id']
```

```python
df.groupby(["director","type"]).size()
```

```python
df.groupby("director").count().nlargest(10, 'show_id')['show_id']
```

```python
df.groupby(["country","type"]).size()
```

```python
df.groupby("country").count().nlargest(20, 'show_id')['show_id']
```

```python
df.groupby(["listed_in","type"]).size()
```

```python
df.groupby("listed_in").count().nlargest(20, 'show_id')['show_id']
```
